```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("titanic.csv")
print(df)
```

```python
print(df.shape)
```

```python
print(df.isna().sum())
```

```python
mean=df['Age'].mean()
print(mean)
```

```python
df['Age']=df['Age'].fillna(mean)
df['Age']
```

```python
newdf = df.drop(columns = ['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'])
print(newdf.to_string())
```

```python
newdf1 = pd.get_dummies(newdf,dtype=int)
print(newdf1.to_string())
```

```python
x = newdf1.drop(columns = ['Survived'])
print(x)
y = df['Survived']
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.20)
model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
ans=model.predict(X_test)
```

```python
n = int(input("Enter No of Passengers : "))

for i in range(n):
  Pclass=int(input("Enter Pclass: "))
  Fare=float(input("Enter Fare: ")) 
  Sex_male = 0
  Sex_female = 0
  age = int(input("Enter Age : "))
  gender = input("Enter Gender (Male or Female) : ").capitalize()
  if gender == "Female":
    Sex_female = 1
  elif gender == "Male":
    Sex_male = 1
    ans = model.predict([[age,Sex_female,Sex_male,Pclass,Fare]])
  if ans == 1:
    print("Passenger Survived")
  else:
    print("Passenger did not Survive")
```

```python
model.score(X_train,Y_train)
```
